```python
#Importing important libraries
import pandas as pd     
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a dictionary with 30 days of dummy data
data = {
    'Date': pd.date_range(start='2024-08-10', periods=30),
    'Open': np.random.uniform(low=50000, high=70000, size=30),
    'High': np.random.uniform(low=50000, high=70000, size=30),
    'Low': np.random.uniform(low=50000, high=70000, size=30),
    'Close': np.random.uniform(low=50000, high=70000, size=30),
    'Volume': np.random.uniform(low=30000, high=200000, size=30),
    'Market Cap': np.random.uniform(low=1000000, high=1100000, size=30)
}
```

```python
# Create DataFrame 
df = pd.DataFrame(data)
print(df)

```

```python
#Applying filter conditions

```

```python
#The `Close` price is greater than the `Open` price leading to a positive gain
positive_gain = df[df['Close'] > df['Open']]
print(positive_gain)
```

```python
#The trading `Volume` was higher than the average trading volume for that period.
average_volume = df['Volume'].mean()
high_volume = df[df['Volume'] > average_volume]
print(high_volume)
```

```python
#For both a positive gain and trading volume greater than average volume.
filtered_df = df[(df['Close'] > df['Open']) & (df['Volume'] > average_volume)]
print(filtered_df)
```

```python
#For Loop and Conditional Statements
```

```python
#Creating list to store values
price_change = []
price_trend = []

```

```python
#Using a for loop to iterate through the loop and calculating daily price changes
for index, row in df.iterrows():   
    change = row['Close'] - row['Open']
    price_change.append(change)
    
   #Apply an if-else condition to create another column `Price Trend`
    if row['Close'] > row['Open']:
        price_trend.append('Up')
    else:
        price_trend.append('Down')


```

```python
#Getting the Price change and Price Trend values
df['Price Change'] = price_change
df['Price Trend'] = price_trend


print(df[['Date', 'Open', 'Close', 'Price Change', 'Price Trend']])
```
